```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_digits,load_sample_image
```

```python
digits = load_digits()
```

```python
X = digits.data
Y = digits.target
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=10)
```

```python
pre = km.fit_predict(X)
```

```python
labels = np.zeros_like(pre)
```

```python
pre.shape
```

```python
labels.shape
```

```python
from scipy.stats import mode
```

```python
for i in range(10):
    mask = (pre==i)
    labels[mask] = mode(Y[mask])[0]
```

```python
labels
```

```python
pre
```

```python
temp1 = np.array([0,0,1,2,1,0,2])
temp2 = np.array([7,7,7,9,8,7,9])
```

```python
var = np.zeros_like(temp1)
```

```python
for i in range(3):
    m = (temp1==i)
    print(m)
    var[m] = mode(temp2[m])[0]
```

```python
m
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(confusion_matrix(temp1,var))
```

```python
print(confusion_matrix(Y,labels))
```

```python
accuracy_score(Y,labels)
```

```python
china = load_sample_image("china.jpg")
plt.imshow(china)
```

```python
x, y, z = china.shape
china_2d = china.reshape(x*y, z)
china_2d.shape
```

```python
kmeans_cluster = KMeans(n_clusters=7)
kmeans_cluster.fit(china_2d)
cluster_centers = kmeans_cluster.cluster_centers_
```
